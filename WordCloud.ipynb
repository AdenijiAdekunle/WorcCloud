## WordCloud

 I've always wondered how words are carved into different shapes, even human face. Took a while before I know there is such as 'wordcloud'... Decided to make mine too!

```python
#import the necessary libraries
import re
import numpy as np
from os import path
from PIL import Image #for image collection
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import sys, os

import matplotlib.pyplot as plt #for visualization
%matplotlib inline
```

```python
#open text and set it to read mode... we also give it a name "words"
words = open('wordcloud.txt', mode='r').read()
```

```python
#we want to set some parameters of our wordCloud and set that to a variable "wordC
wordC = WordCloud(width=3000, height=3000, background_color='white').generate(words)
```

```python
#let's call our words into image and see the magic
wordC.to_image()
```

```python
#we can also save it as we want
wordC.to_file('wordC.png')
```

### Also, we can mask the wordcloud into different shapes to make it more fancier

```python
#to add our desired shape, we will open it with the 'Image' of the 'PIL' library and call it mask_wordC
mask_wordC = np.array(Image.open('email_mask.png'))
```

```python
masked = WordCloud(colormap='coolwarm',mask=mask_wordC).generate(words)
```

```python
masked.to_image()
```

```python
masked.to_file('email.png')
```
